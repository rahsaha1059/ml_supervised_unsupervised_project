```python
def calculate_invoice_value(row):
  if row['Coupon_Status'] == 'Used':
    return (((row['Quantity'] * row['Avg_Price'])*(1-row['Discount_pct'] / 100)) * (1 + row['GST'] / 100)) + row['Delivery_Charges']
  else :
    return ((row['Quantity'] * row['Avg_Price']) * (1 + row['GST'] / 100)) + row['Delivery_Charges']
```

```python
q6_df["Coupon_Used_Flag"] = q6_df["Coupon_Status"].isin(["Used"])
revenue_share = q6_df.groupby("Coupon_Used_Flag")['Invoice_Value'].sum()
revenue_share_percentage = (revenue_share / revenue_share.sum()) * 100
print(revenue_share, revenue_share_percentage)
```

```python
from scipy.stats import ttest_ind
```

```python
coupon_revenue = q6_df.loc[q6_df["Coupon_Used_Flag"], "Invoice_Value"]
coupon_revenue
```

```python
non_coupon_revenue = q6_df.loc[~q6_df["Coupon_Used_Flag"], "Invoice_Value"]
non_coupon_revenue
```

**conclusion question-6:**
1) Very strong evidence that coupon usage changes transaction value since the p-value(**1.4060559467634303e-44**) < 0.05

# question - 10
Segment customers into groups such as Premium, Gold, Silver, and Standard. What targeted strategies can be developed for each segment to improve retention and revenue? (Use RFM segmentation techniques)

```python
joined_df.head(3)
```

```python
from datetime import timedelta
current_date=joined_df['Transaction_Dt'].max() + timedelta(days=1)
```

```python
rfm=joined_df.groupby('CustomerID').agg({'Transaction_Dt':lambda x: (current_date - x.max()).days,'Transaction_ID': 'count' , 'Invoice_Value':'sum'}).reset_index()
```

```python
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
rfm.head(5)
```

```python
plt.hist(rfm['Monetary'])
```

```python
r_bins=[0,100,200,300,np.inf]
f_bins=[0,200,400,600,np.inf]
m_bins=[0,20000,40000,60000,np.inf]
```

```python
rfm['R_Score']=pd.cut(rfm['Recency'],bins=r_bins,labels=False)+1
rfm['F_Score']=pd.cut(rfm['Frequency'],bins=f_bins,labels=False)+1
rfm['M_Score']=pd.cut(rfm['Monetary'],bins=m_bins,labels=False)+1
```

```python
rfm
```

```python
rfm['RFM_Score']=rfm['R_Score']+rfm['F_Score']+rfm['M_Score']
rfm
```

```python
plt.hist(rfm['RFM_Score'])
```

```python
rfm['RFM_Score'].mean()
```

```python
def customer_segmentation(rfm_score):
  if rfm_score >=10:
    return "Premium"
  elif rfm_score >=8:
    return "Gold"
  elif rfm_score >=5:
    return "Silver"
  else :
    return "Standard"
```

```python
rfm['Customer_Category']= rfm['RFM_Score'].apply(customer_segmentation)
```
